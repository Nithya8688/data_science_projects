```python
x=df2.drop(['selling_price'],axis=1)
y=df2['selling_price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
linearmodel=LinearRegression()
linearmodel.fit(x_train,y_train)
```

```python
training_data_prediction = linearmodel.predict(x_train)
```

```python
error_score = metrics.r2_score(y_train, training_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
test_data_prediction = linearmodel.predict(x_test)
error_score = metrics.r2_score(y_test, test_data_prediction)
print("R squared Error : ", error_score)
```

```python
plt.scatter(y_test, test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title(" Actual Prices vs Predicted Prices")
plt.show()
```

```python
Adj_r2 = 1 - (1-metrics.r2_score(y_train, training_data_prediction)) * (len(y)-1)/(len(y)-x.shape[1]-1)
```
